# web cam

````python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if ret == False:
        break
    cv2.imshow('Video Window', frame)
    cv2.imshow('Video Window2', cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))
    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
cap.release()
````

# recording a video

````python
cap  = cv2.VideoCapture(0)                                    # capture the webcam
fourcc = cv2.VideoWriter_fourcc(*'XVID')                      # define the codec
out = cv2.VideoWriter('camera.avi', fourcc, 20.0, (640,480))  # define the output file
while True:                                                   # loop until the user presses q
    ret, frame = cap.read()                                   # read the next frame
    if ret == False:                                          # if the frame is not captured correctly
        break                                                 # break the loop
    out.write(frame)                                          # write the frame to the file
    cv2.imshow('Video Window', frame)                         # show the frame 
    if cv2.waitKey(1) == ord('q'):                            # if the user presses q
        break                                                 # break the loop
cv2.destroyAllWindows()                                       # close all the windows
cap.release()                                                 # release the webcam
out.release()                                                 # release the file              
````
